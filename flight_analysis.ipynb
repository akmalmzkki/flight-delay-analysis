## Weather Dataset

```python
df3 = pd.read_csv('Weather Dataset.csv')
df3
```

```python
# mengecek apakah ada data yang kosong
df3.isnull().sum()
```

```python
# mengubah tipe data atribut AirportID menjadi string
df3['AirportID'] = df3['AirportID'].astype(str)
```

## Penggabungan Dataset Airport codes dengan Dataset Flight on Time

```python
df2 = pd.merge(df2, df1, left_on='OriginAirportID', right_on='airport_id', suffixes=('_Flight', '_Airport'))
df2.drop_duplicates(inplace=True)
df2
```

```python
# Drop kolom airport_id
df2.drop(['airport_id'], axis=1, inplace=True)
df2
```

# Preprocessing

## Preprocessing keterlambatan penerbangan berdasarkan keadaan cuaca

```python
# Mangambil nilai depdelay yang negatif
df2_1 = df2[['OriginAirportID', 'DepDelay']]
df2_1 = df2_1[df2_1['DepDelay'] < 0]
df2_1.drop_duplicates(inplace=True)
df2_1
```

```python
# mengambil atribut airportID dan skycondition
df3_1 = df3[['AirportID', 'SkyCondition']]
df3_1.drop_duplicates(inplace=True)
df3_1
```

```python
df2_1['DepDelay'].dtypes
```

```python
# penggabungan data flight dan weather berdasarkan id yang sama
merged1 = pd.merge(df2_1, df3_1, left_on='OriginAirportID', right_on='AirportID', suffixes=('_flight', '_weather'))
res1 = merged1[['DepDelay', 'SkyCondition']]
res1.drop_duplicates(inplace=True)
res1
```

```python
# Menggabungkan nilai skyCondition berdasrkan rata-rata dari setiap atribut
res1 = res1.groupby(['SkyCondition']).mean().reset_index()
res1
```

```python
# mengubah tipe data SkyCondition menjadi kategori
res1['SkyCondition'] = res1['SkyCondition'].astype('category')
res1.dtypes
```

```python
# Mengambil 3 huruf didepan pada atribut SkyCondition
res1['SkyCondition'] = res1['SkyCondition'].str[:3]
res1
```

```python
# menghitung rata-rata depDelay berdasarkan SkyCondition
res1 = res1.groupby(['SkyCondition']).mean().reset_index()
res1
```

## Preprocessing keterlambatan dengan mempertimbangkan faktor suhu udara dan aliansi penerbangan

```python
# Mengambil data OriginAirportID, DepDelay, dan Carrier yang memiliki nilai depdelay negatif
df2_2 = df2[['OriginAirportID', 'DepDelay', 'Carrier']]
df2_2 = df2_2[df2_2['DepDelay'] < 0]
df2_2.drop_duplicates(inplace=True)
df2_2
```

```python
# mengambil data airportID dan wetbulbfarenheit
df3_2 = df3[['AirportID', 'WetBulbFarenheit']]
df3_2.drop_duplicates(inplace=True)
df3_2
```

```python
# menampilkan carrier yang berbeda
df2_2['Carrier'].unique()
```

```python
# menampilkan tipe data tiap atribut
df3_2.dtypes
```

```python
# mengubah isi dari atribut WetBulbFarenheit yang bernilai M menjadi 0
df3_2['WetBulbFarenheit'] = df3_2['WetBulbFarenheit'].replace('M', 0).astype('float')
df3_2
```

```python
# membuat rata-rata wetbulbfarenheit berdasarkan airportID
df3_2 = df3_2.groupby(['AirportID']).mean().reset_index()
df3_2
```

```python
# Merge data flight dan weather berdasarkan id yang sama
merged2 = pd.merge(df2_2, df3_2, left_on='OriginAirportID', right_on='AirportID', suffixes=('_flight', '_weather'))
res2 = merged2[['OriginAirportID', 'DepDelay', 'Carrier', 'WetBulbFarenheit']]
res2.drop_duplicates(inplace=True)
res2
```

```python
# menghitung rata-rata depDelay dan wetbulbfarenheit berdasarkan carrier
res2 = res2.groupby(['Carrier']).mean().reset_index()
res2 = res2.sort_values(by=['DepDelay'])
res2
```

```python
# Kelompokkan carrier berdasarkan aliansi
res2['Aliance'] = np.where(res2['Carrier'].isin(['US', 'UA']), 'Star Alliance', 
                           np.where(res2['Carrier'].isin(['CO', 'DL']), 'Sky Team', 
                                    np.where(res2['Carrier'].isin(['WN', 'FL']), 'Southwest', 
                                             'Others')))
res2
```

```python
res2 = res2.groupby(['Aliance']).mean().reset_index()
res2
```

## Preprocessing keterlambatan penerbangan berdasarkan waktu terhadap kombinasi arah angin

```python
# mengambil data OriginAirportID, Year, Month, DayofMonth, dan DepDelay
df2_3 = df2[['OriginAirportID', 'Year', 'Month', 'DayofMonth', 'DepDelay']]
df2_3.drop_duplicates(inplace=True)
df2_3
```

```python
# menggabungkan bulan, tanggal, tahun ke dalam satu atribut baru
df2_3['Date'] = df2_3['Year'].astype(str) + '-' + df2_3['Month'].astype(str) + '-' + df2_3['DayofMonth'].astype(str)
df2_3
```

```python
# mengubah tipe data atribut Date menjadi datetime
df2_3['Date'] = pd.to_datetime(df2_3['Date'])
# mengecek tipe data
df2_3.dtypes
```

```python
df2_3 = df2_3[['OriginAirportID', 'Date', 'DepDelay']]
df2_3.drop_duplicates(inplace=True)
df2_3
```

```python
# Ambil dep delay dibawah 0
df2_3 = df2_3[df2_3['DepDelay'] < 0]
df2_3
```

```python
df3_3 = df3[['AirportID', 'WindDirection']]
df3_3.drop_duplicates(inplace=True)
df3_3
```

```python
df3_3['WindDirection'] = df3_3['WindDirection'].replace('VR ', 0)
df3_3['WindDirection'] = df3_3['WindDirection'].replace('M', 0)
```

```python
# mengubah tipe data atribut WindDirection menjadi float
df3_3['WindDirection'] = df3_3['WindDirection'].astype('float')
df3_3.dtypes
```

```python
# hitung rata-rata wind direction berdasarkan airport id
df3_3 = df3_3.groupby(['AirportID']).mean().reset_index()
df3_3
```

```python
# menggabungkan data flight dan weather berdasarkan id yang sama
merged3 = pd.merge(df2_3, df3_3, left_on='OriginAirportID', right_on='AirportID', suffixes=('_flight', '_weather'))
res3 = merged3[['OriginAirportID', 'Date', 'WindDirection', 'DepDelay']]
res3.drop_duplicates(inplace=True)
res3
```

```python
# Grupkan berdasrkan date
res3 = res3.groupby(['Date']).mean().reset_index()
res3
```

```python
# mengubah tipe data atribut Date menjadi datetime
res3['Date'] = pd.to_datetime(res3['Date'])
res3
```

```python
# Membuat bins Date setiap kenaikan 5 hari
bins = pd.date_range(start='2011-10-01', end='2011-10-31', freq='6D')
labels = [str(bins[i]) + ' -- ' + str(bins[i+1]) for i in range(len(bins)-1)]

res3['Date bins'] = pd.cut(res3['Date'], bins=bins, labels=labels)
res3
```

```python
# Mengubah tipe data atribut Date bins menjadi string
res3['Date bins'] = res3['Date bins'].astype(str)
res3
```

```python
# Menghapus nilai jam pada Date bins serta mengganti nilai nan dengan 2011-10-01--2011-10-06
res3['Date bins'] = res3['Date bins'].str.replace('00:00:00', '')
res3['Date bins'] = res3['Date bins'].str.replace(' ', '')
res3['Date bins'] = res3['Date bins'].str.replace('nan', '2011-10-01--2011-10-07')
res3
```

```python
# Menggabungkan Date bins berdasarkan rata-rata wind direction dan dep delay
res3 = res3.groupby(['Date bins']).mean().reset_index()
res3
```

## Preprocessing keterlambatan penerbangan berdasarkan waktu keberangkatan dan kecepatan angin

```python
# Mengambil data yang hanya berhubungan dengan airport code
df2_4 = df2[['OriginAirportID', 'CRSDepTime', 'DepDelay']]
df2_4.drop_duplicates(inplace=True)
df2_4
```

```python
# Mengambil data AirportID dan WindSpeed
df3_4 = df3[['AirportID', 'WindSpeed']]
df3_4.drop_duplicates(inplace=True)
df3_4
```

```python
# Merge data flight dan weather berdasarkan id yang sama
merged4 = pd.merge(df2_4, df3_4, left_on='OriginAirportID', right_on='AirportID', suffixes=('_df1', '_df3'))
res4 = merged4[['CRSDepTime', 'WindSpeed', 'DepDelay']]
res4.drop_duplicates(inplace=True)
res4
```

```python
# Tampilkan data dengan DepDelay < 0
res4 = res4[res4['DepDelay'] < 0]
res4
```

```python
res4['WindSpeed'] = res4['WindSpeed'].str.replace(' ', '0').str.replace('M', '0').astype('float')
res4.dtypes
```

```python
# Grupkan berdasarkan OriginAirportID
res4 = res4.groupby(['CRSDepTime']).mean().reset_index()
res4.sort_values(by=['DepDelay'])
```

```python
# Mengubah tipe data CRSDepTime menjadi string
res4['CRSDepTime'] = res4['CRSDepTime'].astype('str')
```

```python
# Jika panjang nilai dari atribut = 2, maka tambahkan angka 00 di depannya
res4['CRSDepTime'] = res4['CRSDepTime'].apply(lambda x: '00' + x if len(x) == 2 else x)
res4['CRSDepTime'] = res4['CRSDepTime'].apply(lambda x: '0' + x if len(x) == 3 else x)
res4
```

```python
# Menambahkan . pada CRSDepTime
res4['CRSDepTime'] = res4['CRSDepTime'].apply(lambda x: x[:2] + ':' + x[2:])
res4
```

```python
# Mengubah tipe data CRSDepTime menjadi time dengan hanya menampilkan jam dan menit saja
res4['CRSDepTime'] = pd.to_datetime(res4['CRSDepTime'], format='%H:%M')
res4
```

```python
# Membuat bins berdasrkan CRSDepTime dengan setiap kenaikan 3 jam, contohnya 00:00 - 03:00, 03:00 - 06:00, dst
bins = pd.date_range('1900-01-01 00:00:00', '1900-01-01 23:59:00', freq='3H')
labels = [f'{i} - {j}' for i, j in zip(bins[:-1].strftime('%H:%M'), bins[1:].strftime('%H:%M'))]
res4['CRSDepTime bin'] = pd.cut(res4['CRSDepTime'], bins=bins, labels=labels, right=False)
```

```python
# groupkan berdasarkan CRSDepTime bin
res4 = res4.groupby(['CRSDepTime bin']).mean().reset_index()
res4
```
